```python
#importing pandas and numpy
import pandas as pd
import numpy as ny
```

```python
#import normal data -nn- , making the index and drop gene ID
df1=pd.read_table("nn.txt")
df1=df1.set_index("Hugo_Symbol")
df1=df1.drop(["Entrez_Gene_Id"],axis=1)
```

```python
#import cancer data -mm- , making the index and drop gene ID
df2=pd.read_table("mm.txt")
df2=df2.set_index("Hugo_Symbol")
df2=df2.drop(["Entrez_Gene_Id"],axis=1)
```

```python
#making the data in organize  tabels
data11=pd.DataFrame(df1)
data22=pd.DataFrame(df2)
```

```python
 #df1 and df1 as numpy array
datah =ny.array(df1)
datac =ny.array(df2)
#copy the data
datahh=datah
datacc=datac
```

```python
#git the index of genes from cancer data that is corrupt stage 2
oo=[]
for i in range (19648):
    t=0
    ss=datah[i]
    for x in range (50):
        if ss[x]==0 :
            t=t+1
            
    if t>25:
        oo.append(i)
```

```python
#get the number or genes that will be deleted in stage 1
len (oo)
```

```python
#delete corrupt gene from normal data  stage 1 
new_datah=ny. delete(datah, oo , 0)
```

```python
#delete corrupt gene from cancer  data  stage 1
new_datac=ny. delete(datac, oo, 0)
```

```python
#git the index of genes from cancer data that is corrupt stage 2
oo2=[]
for i in range (17675):
    t=0
    ss=new_datac[i]
    for x in range (50):
        if ss[x]==0 :
            t=t+1
            
    if t>25:
        oo2.append(i)
```

```python
#get the number or genes that will be deleted in stage 2
len (oo2)
```

```python
#delete corrupt gene from normal data  stage 2 
datah2=ny. delete(new_datah, oo2, 0)
```

```python
#delete corrupt gene from cancer data  stage 1 
datac2=ny. delete(new_datac, oo2, 0)
```

```python
#make our data numpy arrar
datah =ny.array(datah2)
datac =ny.array(datac2)
```

```python
len(datah)
```

```python
#calculate the pearsonr 's correlation between two variables deu to the linearity of data
from scipy.stats import pearsonr
#make array to take the correlation coeff
r=[]
for i in range (17337):
    # prepare data
    data_hy = datah[i]
    data_ca = datac[i]
    # calculate Pearson's correlation
    corr ,_=pearsonr (data_hy,data_ca)
    r.append(corr)


    
```

```python
#sorting the correlation array
sortcorr=ny.sort(r)
print(sortcorr)
```

```python
#min -ve cc
mincorr=min(r)
print ('min -ve cc : '+str(mincorr))
```

```python
#max +ve cc
maxcorr=max(r)
print("max +ve cc : "+str(maxcorr))
```

```python
#if we want calculate the pearsonr 's correlation between two variables to get the index
from scipy.stats import pearsonr
#make array to take the correlation coeff
u=[]
for i in range (19648):
    # prepare data
    data_h = datahh[i]
    data_c = datacc[i]
    # calculate Pearson's correlation
    corr ,_=pearsonr(data_h,data_c)
    u.append(corr)
```

```python
#getting the index for highest +ve cc
maxindex=u.index(max(r))
print ("the index  highest +ve cc : "+str(maxindex))
```

```python
#getting the index for the lowest -ev cc
minindex=u.index(min(r))
print("the index for the lowest -ve cc : " +str(minindex))
```

```python
#printing the highest  +ve cc in normal data
data11[maxindex:maxindex+1]
```

```python
#printing the highest  +ve cc in cancer data (for check)
data22[maxindex:maxindex+1]
```

```python
#printing the lowest -ve cc in normal data
data11[minindex:minindex+1]
```

```python
#printing the lowest -ve cc in cancer data (for check)
data22[minindex:minindex+1]
```

```python
#plotting the expression level for the highest +ve cc
#importing pyplot
from matplotlib import pyplot
# prepare data
data1 = datahh[maxindex]
data2 = datacc[maxindex]
# plot
pyplot.scatter(data1, data2)
pyplot.show()
```
